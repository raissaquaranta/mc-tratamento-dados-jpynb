```python
Y = np.linspace(4.0,0,400)

sns.set_theme()
fig,axs = plt.subplots(1, figsize = (20, 14))
plt.suptitle('Decaimento em Profundidade da Fluência', y=0.94, x=0.51, fontsize=30)
plt.subplots_adjust(hspace = 0.7)   # Pushing up the title
axs.set_title('$10^8$ fótons - 5 simulações - 30 mW/mm²', fontsize=20)

# Actual Lateral View Plot
axs.plot(Y, 30*0.00785398*data_50_s[:,400], color = 'blue', label = '0.5 mm',linewidth=4.0)
axs.plot(Y, 30*0.03141592*data_100_s[:,400], color = 'red', label = '1.0 mm',linewidth=4.0)
axs.plot(Y, 30*0.1256637*data_200_s[:,400], color = 'green', label = '2.0 mm',linewidth=4.0)
plt.legend(loc=1, prop={'size': 25})

# # Legend for the axis
axs.set_xlabel('Profundidade [mm]', fontsize=30)
plt.xticks(fontsize=25)
plt.yticks(fontsize=25)
axs.set_ylabel('Fluência Real [mJ/mm²]', fontsize=30)

plt.savefig('grafico3-decaimento-raios-reais-30mW.jpg', dpi=200)
plt.show()
```
